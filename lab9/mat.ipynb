```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
bank=pd.read_csv("D:/christ/py lab/lab9/bank.csv")
print(bank.info)
```

```python
bank['job'].unique()
# bank['balance'].unique()
```

```python
print(bank.groupby('job').size())
```

```python
print(bank.groupby('job').min())
```

```python
selfemployed=bank[bank['job']=='self-employed']
service=bank[bank['job']=='services']
entrepreneurs=bank[bank['job']=='entrepreneur']
```

```python
bank["age"].plot(kind="hist",xlabel="age",color='b',fontsize=10,title="Age Distribution Histogram")
```

```python
Type=list(bank['job'].unique())
count=list(bank['job'].value_counts())
Avg=list(bank.groupby('job')["balance"].mean())
plt.bar(Type,Avg,color=['maroon','yellow','red'],width=0.4)
plt.xlabel('Job Categories')
plt.ylabel('Counts')
plt.title('Job Categories Bar Graph')
plt.show()
```

```python
bank.plot(x="age",y="balance",kind="scatter",label='setosa',color='b')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.title('Age vs. Balance Scatterplot')
plt.grid(True)
plt.show()
```

```python
bank["balance"].plot(kind="line",label='setosa',color='b',linewidth='1.5',ls=':')
plt.axis([0, 150, 2, 5])
plt.title('line plot')
plt.xlabel('SepalWidthCm')
plt.show()
```

```python
selfemployed["balance"].plot(kind="line",label='setosa',color='r',linewidth='1.5',ls=':')
service["balance"].plot(kind="line",label='setosa',color='b',linewidth='1.5',ls=':')
entrepreneurs["balance"].plot(kind="line",label='setosa',color='g',linewidth='1.5',ls=':')
plt.title('job wise-line plt')
plt.xlabel('sepalwidthcm')
plt.legend()
plt.show()
```

```python
entrepreneurs.plot.box()
plt.title("box plot")
plt.legend()
plt.show()
```

```python
selfemployed.plot.area()
plt.title("box plot")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = bank['marital'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
plt.title('martial status')
plt.axis('equal') 
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


columns_to_include = ['age', 'balance', 'duration', 'campaign']

sns.pairplot(bank[columns_to_include], diag_kind='kde')  # 'diag_kind' for diagonal plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

import pandas as pd
data = pd.DataFrame({'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
                     'Value': [15, 30, 10, 45, 25, 35]})

sns.violinplot(x='Category', y='Value', data=data)
plt.title('Violin Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd


data = pd.DataFrame({'A': [1, 2, 3, 4, 5],
                     'B': [2, 3, 4, 5, 6],
                     'C': [3, 4, 5, 6, 7],
                     'D': [4, 5, 6, 7, 8],
                     'E': [5, 6, 7, 8, 9]})


corr_matrix = data.corr()  
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
